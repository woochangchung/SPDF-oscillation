```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lat_dat = np.load("generic-lattice-parameters.npz")
V0, tunneling, Us = [lat_dat[x] for x in lat_dat]
```

```python
#1,1 & 1,-1 pair
UAA = Us * 100.4 / 100.
UAB = Us * 101.333 / 100.
u_Hz_n = UAA - UAB
J_Hz_n = 4*tunneling**2 / UAB

#1,-1 & 1,0 pair
UAA = Us * (100.4 + 100.867) / 2. / 100.
UAB = Us * 100.4 / 100.
u_Hz_p = UAA - UAB
J_Hz_p = 4*tunneling**2 / UAB
```

```python
fig, ax = plt.subplots(2, 1, figsize = (3.2, 3), sharex = True)
fig.subplots_adjust(hspace=0)

ax[0].plot(V0, J_Hz_n, label = r"$|a\rangle = |1, -1\rangle$, $|b\rangle = |1,1\rangle$")
ax[0].plot(V0, J_Hz_p, ls = "--", label = r"$|a\rangle = |1, -1\rangle$, $|b\rangle = |1,0\rangle$")
ax[0].set_ylabel("$J/2\pi$ (Hz)")
ax[0].legend(fontsize = 8)
ax[0].tick_params(labelsize = 8)

ax[1].plot(V0, u_Hz_n)
ax[1].plot(V0, u_Hz_p, ls = "--")
ax[1].set_xlabel("Longitudinal lattice depth ($E_r$)")
ax[1].set_ylabel("$D/2\pi$ (Hz)")
ax[0].tick_params(labelsize = 8)

plt.tight_layout()
fig.savefig("J-and-D-vs-lattice-depth.pdf")
```

```python
fig, ax = plt.subplots(2, 1, figsize = (3.2, 3), sharex = True)
fig.subplots_adjust(hspace = 0)

ax[0].plot(V0, J_Hz_n, label = r"$|a\rangle = |1, -1\rangle$, $|b\rangle = |1,1\rangle$")
ax[0].plot(V0, J_Hz_p, ls = "--", label = r"$|a\rangle = |1, -1\rangle$, $|b\rangle = |1,0\rangle$")
ax[0].set_ylabel("$J/2\pi$ (Hz)")
ax[0].legend(fontsize = 8)
ax[0].tick_params(labelsize = 8)

ax[1].semilogy(V0, -u_Hz_n/J_Hz_n)
ax[1].semilogy(V0, u_Hz_p/J_Hz_p, ls = "--")
ax[1].set_xlabel("Longitudinal lattice depth ($E_r$)")
ax[1].set_ylabel("$D/J$", fontsize = 10)
ax[1].tick_params(labelsize = 8)

plt.tight_layout()
fig.savefig("J-and-DoJ-vs-lattice-depth_log.png")
```
